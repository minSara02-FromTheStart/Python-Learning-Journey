```python
import pandas as pd

income = [10000, 12000, 15000, 18000, 20000, 500000]
income_series = pd.Series(income)
```

```python
income_series.mean()
```

```python
income_series.median()

```

```python
income_series.mode()
```

```python
income_series.std()
```

```python
income.append(1000000)
income_series = pd.Series(income)
income_series.mean()
income_series.median()

```

```python
print("Mean income:", income_series.mean())
print("Median income:", income_series.median())
print("Mode income:", income_series.mode())
```

```python
print("Variance:", income_series.var())
print("Standard Deviation:", income_series.std())
```

```python
income.append(1000000)
income_series = pd.Series(income)

print("New Mean:", income_series.mean())
print("New Median:", income_series.median())
```

```python
income = [100,120,150,180,500]
income_series = pd.Series(income)
print("income_series.describe()=\n", income_series.describe())
print("income_series.mean() = ", income_series.mean())
print("income_series.median() = ", income_series.median())
print("income_series.mode() = ", income_series.mode())
```
