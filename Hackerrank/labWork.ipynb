```python

#Problem 1: Student Database
#Task 
#Create a dictionary where keys are student IDs (101 to 110) and values are dictionaries 
#containing student information (Name, CGPA, Credits) 
#Requirements 
# ● Use a loop to generate IDs from 101 to 110 
# ● Name should be "Student1" for ID 101, "Student2" for ID 102, etc. 
# ● CGPA should be 3.0 + (id-100)/10 (so 101→3.1, 102→3.2, etc.) 
# ● Credits should be 50 for all students 
# ● Print the information for student ID "105"

student_database = {}   # create once

for i in range(101, 111):
    student_id = i
    name = "Student" + str(i - 100)
    cgpa = 3.0 + (i - 100) / 10
    credits = 50

    profile = {"Name": name, "CGPA": cgpa, "Credits": credits}
    student_database[student_id] = profile

print(student_database[105])
```
