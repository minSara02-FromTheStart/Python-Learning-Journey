```python

#Problem 1: Student Database
#Task 
#Create a dictionary where keys are student IDs (101 to 110) and values are dictionaries 
#containing student information (Name, CGPA, Credits) 
#Requirements 
# ● Use a loop to generate IDs from 101 to 110 
# ● Name should be "Student1" for ID 101, "Student2" for ID 102, etc. 
# ● CGPA should be 3.0 + (id-100)/10 (so 101→3.1, 102→3.2, etc.) 
# ● Credits should be 50 for all students 
# ● Print the information for student ID "105"

student_database = {}   # create once

for i in range(101, 111):
    student_id = i
    name = "Student" + str(i - 100)
    cgpa = 3.0 + (i - 100) / 10
    credits = 50

    profile = {"Name": name, "CGPA": cgpa, "Credits": credits}
    student_database[student_id] = profile

print(student_database[105])
```

```python
#Problem 2: Grade Statistics with Tuples
#Task
#Given a list of student grades, create a function that returns a tuple containing statistics about
#the grades.
#Requirements
#● Write a function called grade_stats(grades)
#● Return a tuple with three values: (highest_grade, lowest_grade, average_grade)
#● Print the tuple
#Given Data
#grades = [85, 92, 78, 90, 88, 76, 95, 89]
def grade_stats(grades):
    highest_grade = max(grades)
    lowest_grade = min(grades)
    average_grade = sum(grades)/len(grades)
    return(highest_grade,lowest_grade,average_grade)

grades = [85, 92, 78, 90, 88, 76, 95, 89]
result = grade_stats(grades)
print(result)
```

```python
#Problem 3: Unique Course Enrollment
#Task
#You have enrollment data from multiple sources. Find unique courses and common courses
#between semesters.
#Given Data
#fall_courses = {"CSE110", "CSE111", "MAT101", "ENG101", "CSE110"} # duplicate CSE110
#spring_courses = {"CSE111", "CSE246", "MAT102", "PHY101"}
#Requirements
#● Create a set of unique fall courses (remove duplicates)
#● Find courses offered in BOTH semesters (intersection)
#● Find courses ONLY in fall semester (difference)
#● Find ALL unique courses across both semesters (union)
#● Add "CSE325" to the spring courses set

fall_courses = {"CSE110", "CSE111", "MAT101", "ENG101", "CSE110"}
spring_courses = {"CSE111", "CSE246", "MAT102", "PHY101"}
unique_fall = fall_courses
print("Unique Fall Courses:", unique_fall)
both_semesters = fall_courses.intersection(spring_courses)
print("Courses in both semester: ", both_semesters)
only_fall = fall_courses.difference(spring_courses)
print("Courses only in Fall semester : ", only_fall)
all_unique = fall_courses.union(spring_courses)
print("All unique coursses across both semesters: ", all_unique)
spring_courses.add("CSE325")
print("New spring courses set: ", spring_courses)
```
