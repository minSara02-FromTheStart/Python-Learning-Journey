```python
#TASK 1: Variables, Print, Input (Intermediate)
#Practice 1: Student Report Generato
#Task
#Take input: name, id, marks in 3 subjects
#Calculate:
#total
#average
#grade (A/B/C/F)
#Print a formatted report
#Concepts used
#Variables
#Input
#Print formatting (f-string)

name = input("Enter student name: ")
std_id = input("Enter student ID: ")
marks1 = float(input("Enter marks in subject 1: "))
marks2 = float(input("enter marks in subject 2: "))
marks3 = float(input("Eneter marks in subject 3: "))
total = marks1 + marks2 + marks3
avr = total/3
print("Total marks in 3 subjects: ", total)
print("Average marks: ", avr)
if avr >= 80:
    grade = 'A'
elif avr >= 60:
    grade = 'B'
elif avr >= 40:
    grade = 'C'
else:
    grade = 'F'
print(f"student name: {name}")
print(f"Student ID: {std_id}")
print(f"Total Marks: {total}")
print(f"Average Marks: {avr}")
print(f"Grade: {grade}")
```
